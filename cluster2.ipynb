```python
import pandas as pd
data = pd.read_excel('/Users/apple/code/python/homework/cluster/3.xlsx')
data
```

```python
data.columns
```

```python
data=data[['开航日期','客户','承运人','实装箱量','应收合计','应付合计']]
data
```

```python
data_temp=data.groupby(['开航日期','客户','承运人','实装箱量']).sum()
data_group = pd.DataFrame(data_temp).reset_index()
data_group
```

```python
from sklearn.cluster import KMeans
# model = KMeans(n_clusters = 4, n_jobs = 4, max_iter = 4) #分为k类，并发数4，默认8类300轮
model = KMeans(n_clusters = 8)
y_pred=model.fit_predict(data_group[['应付合计','应收合计']]) #开始聚类
```

```python
import matplotlib.pyplot as plt
# 允许在笔Notebook中输出绘图
%matplotlib inline 
# 使图具有更好的质量
%config InlineBackend.figure_format = 'svg'
# 设置默认样式
plt.style.use("ggplot") 
#支持显示中文标签 
plt.rcParams['font.sans-serif']=['SimHei']
plt.scatter(data_group['应付合计'], data_group['应收合计'], c=y_pred)
plt.xlabel('应付合计') # 设置横轴标签
plt.ylabel('应收合计') # 设置纵轴标签
plt.show()
```
